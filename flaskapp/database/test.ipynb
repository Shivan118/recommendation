```python
df = jobs_df[['Name_of_employer', 'User_id']]
```

```python
df['assigned'] = 0
```

```python
(df.loc[df['Name_of_employer'] == 'DRP P', 'assigned'].iloc[0] == 0)
```

```python
temp = 1
for index, row in df.iterrows():
    row = row.copy()
    if  df.iloc[:index+1].loc[df['Name_of_employer'] == row['Name_of_employer'], 'assigned'].iloc[0] == 0:
        row['User_id'] = temp
        df.loc[index, 'User_id'] = temp
        row['assigned'] = 1
        df.loc[index, 'assigned'] = 1
        temp += 1
    else:
        row['User_id'] = df.iloc[:index+1].loc[df['Name_of_employer'] == row['Name_of_employer'], 'User_id'].iloc[0]
        df.loc[index, 'User_id'] = row['User_id']
        row['assigned'] = 1
        df.loc[index, 'assigned'] = 1
```
